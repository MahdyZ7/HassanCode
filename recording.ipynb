```python
import psutil
import csv

cpu_util_list = []
cpu_freq_list = []
temp_list = []
timing_list = []


def get_cpu_util():
    # Getting CPU Utilization in percentage
    cpu_util = psutil.cpu_percent()
    # add the value to the data array
    cpu_util_list.append(cpu_util)


def get_cpu_freq():
    # Getting CPU Frequency in MHz
    cpu_freq = psutil.cpu_freq().current
    # add the value to the data array
    cpu_freq_list.append(cpu_freq)


def get_cpu_temp():
    # Getting CPU Temperature
    x = psutil.sensors_temperatures()
    temp = x['cpu_thermal'][0].current
    # add the value to the data array
    temp_list.append(temp)


def get_timing(timing):
    # add the value to the data array
    timing_list.append(round(timing, 2))


def timings_normal(x, timings_file):
    # Calculate the overall timing for CNN
    samples = x
    n = len(timing_list)
    duration = timing_list[n - 1]
    avg = duration / samples
    # Opening the csv file and preparing the table
    with open(timings_file, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["algorithm", "operation", "duration", "samples", "avg per sample"])
        # Writing the data in the csv file
        writer.writerow(["ViT", "Inference_full_MHA", duration, samples, avg])


def timings_half(x, timings_file):
    # Calculate the overall timing for CNN
    samples = x
    n = len(timing_list)
    duration = timing_list[n - 1]
    avg = duration/samples
    with open(timings_file, "a", newline="") as file:
        writer = csv.writer(file)
        # Writing the data in the csv file
        writer.writerow(["ViT", "Inference_half_MHA", duration, samples, avg])


def record_normal(cpu_file):
    with open(cpu_file, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Benchmark", "Utilization", "Frequency", "Temperature", "timing"])
        # Writing the data in the csv file
        for i in range(len(cpu_util_list)):
            writer.writerow(["ViT_full_MHA", cpu_util_list[i], cpu_freq_list[i], "temp_list[i]", timing_list[i]])


def record_half(cpu_file):
    # Opening the csv file and preparing the table
    with open(cpu_file, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Benchmark", "Utilization", "Frequency", "Temperature", "timing"])
        # Writing the data in the csv file
        for i in range(len(cpu_util_list)):
            writer.writerow(["ViT_half_MHA", cpu_util_list[i], cpu_freq_list[i], "temp_list[i]", timing_list[i]])


```
